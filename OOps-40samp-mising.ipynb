```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from matplotlib.backends.backend_pdf import PdfPages
import scipy.cluster.hierarchy as spc
from scipy.stats import pearsonr
from sklearn import metrics


def getDataFrame():
    df = pd.read_csv("Stats_task2_40samp.csv")
    return df


class Data:
    def __init__(self,ID, mz, rt, intensity):
        self.ID = ID
        self.mz = mz
        self.rt = rt
        self.i = intensity
    
    def calculateMeanIntensity(self):
        self.meanIntensity = sum(self.i)/len(self.i)

    def calculateMedIntensity(self):
        self.medIntensity = np.median([x for x in self.i])
        
    def Annotate(self, string):
        self.anno = string
        
    def bin_Num(self, x):
        self.bin = x
        
    def clust_Num(self, x):
        self.cluster = x
        
    def isotope(self, string):
        self.iso = string


def getData():
    df = getDataFrame()
    dfValues = df.values.tolist()
    data = list()
    for i in range(len(dfValues)):
        data.append(Data(dfValues[i][0], dfValues[i][1], dfValues[i][2], [j for j in dfValues[i][12:52]]))
    return data

def create_df(i,j):
    temp = list()
    col = list()
    for x in range(i,j):
        temp.append(data[x].i)
        col.append(x)
    df = pd.DataFrame(temp)
    df.set_axis(col, axis=0, inplace=True)
    return df.T



if __name__ == "__main__":
    data = getData()
    for i in range(len(data)):
        data[i].calculateMeanIntensity()
#     print(data[1].meanIntensity)
#     print(data[2].i)
    
    flag = 0
    # Counting number of zero intensity 
    # and removing feature with greater than 30% zeroes
    
    i=0
    while(i<len(data)):
        cnt=0
        for j in range(0,40):
            if data[i].i[j]==0:
                cnt = cnt+1
        if cnt>12:
            del data[i]
        else:
            i = i+1
            
    
    #replace missing value with median intensity
    # and taking log transformation
    for i in range(len(data)):
        data[i].calculateMedIntensity()
        
    
    for i in range(len(data)):
        for j in range(0,40):
            data[i].i[j] = np.log10(1 + data[i].i[j])
            if data[i].i[j] == 0:
                data[i].i[j] = np.nan

                
    
    #Isotope Detection            
    def isotope(*ind):
        q = 0
        rt_tol = 0.05
        i = ind[0][0]
        j = ind[-1][-1]
        
        for x in range(i,j): 
            for y in range(x+1,j+1):
                d1 = data[x].i
                d2 = data[y].i
                
                correlation = np.corrcoef(d1, d2)[0][1]

                if (1.003-0.01<= abs(data[x].mz-data[y].mz) <=1.003+0.01) and (abs(data[x].rt-data[x].rt)<rt_tol) and correlation>0.4 and data[x].cluster == data[y].cluster:
                    print("Isotopes are m/z %d and %d and corr = %f" %(data[x].mz,data[y].mz, correlation)) 
                    if(data[x].mz>data[y].mz):
                        data[x].isotope("i{q}+1".format(q=q))
                        data[y].isotope("i{q}".format(q=q))
                        q = q + 1
                    else: 
                        data[y].isotope("i{q}+1".format(q=q))
                        data[x].isotope("i{q}".format(q=q))
                        q = q + 1
                    
    #Calculating correalion matrix and forming clusters                
    def corr(*ind):
        i = ind[0][0]
        j = ind[-1][-1]
        df_small = create_df(i,j+1)
        print(df_small.T)

        correlation_mat = df_small.corr()
        sns.heatmap(correlation_mat, annot = True)
        plt.show()
        corelat = correlation_mat .values
        pdist = spc.distance.pdist(corelat)
        linkage = spc.linkage(pdist, method='average')
        t=0
        maxval = -1
        x = 0.01
        
        while(x<=1):
            try:
                idx = spc.fcluster(linkage, x * pdist.max(), 'distance')
#                 print(idx)
                val = metrics.silhouette_score(df_small.T, idx, metric='euclidean')
                if(val>maxval):
                    maxval = val
                    t = x
#                 print(val)
                x = x + 0.01
            
            except:
                idx = spc.fcluster(linkage, x * pdist.max(), 'distance')
                x = x + 0.01
                
        idx = spc.fcluster(linkage, t * pdist.max(), 'distance')
        print("Cluster - ",idx)
        print("Number of cluster = %d" %idx.max())
        
        #Annotation
        
        col = df_small.columns
        t = 0
        
        for q in range(0, len(col)):
            data[col[q]].clust_Num(idx[q])
        
        for i in range(1,idx.max()+1):
            clust = []
            cnt = 0
            for j in range(0,len(idx)):
                if idx[j] == i:
                    clust.append(col[j])
                    cnt = cnt + 1;
            print("cnt = %d"%cnt)
            print(clust)
            
            if(cnt>1):
                maxi = 0
                for x in range(0,len(clust)):
                    if(data[clust[x]].medIntensity>maxi):
                        maxi = data[clust[x]].medIntensity
                        maxind = clust[x]
                

                neutral_mass = data[maxind].mz - 1.007
                
                flag = 0

                #Test for dimer present
                dimer_mass = 2*neutral_mass + 1.007
                for x in range(0,len(clust)):
                    if(abs(data[clust[x]].mz-dimer_mass) <=0.01):
                        data[clust[x]].Annotate('2M{t}+H'.format(t=t))
                        flag = 1
                        print(data[clust[x]].anno)
                        
                # Test for annotation file
                for i in range(1, len(Ano_mass)):
                    if Ano_charge[i]==0:
                        Ano_charge[i] = 1
                    cal_mz = (neutral_mass + Ano_mass[i])/(np.abs(Ano_charge[i]))
                    for x in range(0,len(clust)):
                        if(abs(data[clust[x]].mz-cal_mz) <=0.01):
                            data[clust[x]].Annotate('M{t}+'.format(t=t)+ Ano[i])
                            flag = 1
                            print(data[clust[x]].anno)  
                
                if(flag==1):
                    data[maxind].Annotate('M{t}+H'.format(t=t))
                    print(data[maxind].anno)
                    
            t = t+1
                
    #Binning
    rt_tol = 0.03
    j = 1
    ind = []
    mass = []
    x = 1
    q = 0
    for i in range(800):
        if data[j].rt - data[i].rt >= rt_tol:
            ind.append(i)
            mass.append(data[i].mz)
            data[i].bin_Num(x)
            x = x+1
            print("Bin - ",ind)
            print("Binmass -",mass)
            if len(ind)>1:
                corr(ind)
                isotope(ind)
            else:
                data[i].clust_Num(1)
            ind = []
            mass = []

        else:
            ind.append(i)
            mass.append(data[i].mz)
            data[i].bin_Num(x)

        j = j+1
        
```
